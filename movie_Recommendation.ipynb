```python
#Creating UserPreferences
userPreferences = [
    {'title':'American President, The', 'rating':5},
            {'title':'Tom and Huck', 'rating':3.5},
            {'title':'Heat', 'rating':2},
            {'title':"Waiting to Exhale", 'rating':5},
            {'title':'Toy Story', 'rating':4.5},
            {'title':"Dracula: Dead and Loving It", 'rating':2.5}
            
]
ratedMovies = pd.DataFrame(userPreferences)
ratedMovies
```

```python
#flitering movies by title
ratedId = movies_df[movies_df['title'].isin(ratedMovies['title'].tolist())]
#merging
ratedMovies = pd.merge(ratedId, ratedMovies)

ratedMovies
```

```python
#filtering the movies watched by users that the input has watched 
similarUser = ratings_df[ratings_df['movieId'].isin(ratedMovies['movieId'].tolist())]
similarUser.head()
```

```python
#Groupby creates several sub dataframes where they all have the same value in the column specified as the parameter
similarUserGroup = similarUser.groupby(['userId'])
```

```python
similarUserGroup = similarUserGroup[0:100]
```

```python
#storing the pearson correlation in a dictionary
pearsonCorrelationDict = {}

#for every user group in subset
for name, group in similarUserGroup:
    
    #sorting the input and current user
    group = group.sort_values(by='movieId')
    ratedMovies = ratedMovies.sort_values(by='movieId')
    
    #Getting the n (total similar movies watched) for the formula
    nRatings = len(group)
    
    #Getting the review scores for the movies that they both have in common
    t_df = ratedMovies[ratedMovies['movieId'].isin(group['movieId'].tolist())]
    
    #storing them in a temporary buffer variable in a list format
    tRatingList = t_df['rating'].tolist()
    tGroupList = group['rating'].tolist()
    Sxx = sum([i**2 for i in tRatingList]) - pow(sum(tRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tGroupList]) - pow(sum(tGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tRatingList, tGroupList)) - sum(tRatingList)*sum(tGroupList)/float(nRatings)
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/np.sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
pearsonCorrelationDict.items()
```

```python
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
pearsonDF.head()
```

```python
pearsonDF.columns = ['similarityIndex']
pearsonDF['userId'] = pearsonDF.index
pearsonDF.index = range(len(pearsonDF))
pearsonDF.head()
```

```python
topUser=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50]
topUser.head()
```

```python
topUserRating = topUser.merge(ratings_df, left_on='userId', right_on='userId', how='inner')
topUserRating.head()
```

```python
#multiplying the similarity by the user's rating 
topUserRating['weightedRating'] = topUserRating['similarityIndex']*topUserRating['rating']
topUserRating.head()
```

```python
#Applying a sum to the topUsers after grouping it up by userId
topUsersRating = topUserRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
topUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
topUsersRating.head()
```

```python
#creating an empty dataframe
recommendation_df = pd.DataFrame()
#Now we take the weighted average
recommendation_df['Average recommendation score'] = topUsersRating['sum_weightedRating']/topUsersRating['sum_similarityIndex']
recommendation_df['movieId'] = topUsersRating.index
recommendation_df.head()
```

```python
recommendation_df = recommendation_df.sort_values(by='Average recommendation score', ascending=False)
recommendation_df.head(10)
```
